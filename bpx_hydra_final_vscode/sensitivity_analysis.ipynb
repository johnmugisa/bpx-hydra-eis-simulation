```python
import copy
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

from bpx_hydra_config import *
from bpx_hydra_parameters import (
    make_physical_params_from_bpx,
    make_grouped_params_from_physical,
    run_grouped_spme_eis,
)


# =============================================================================
# SENSITIVITY SETTINGS
# =============================================================================

FACTOR = 2.0
N_VALUES = 11

# For sensitivity plots, a narrower high-frequency limit often gives clearer
# presentation-quality Nyquist curves. Change SENS_FMAX to FMAX if you want
# the full frequency range from bpx_hydra_config.py.
SENS_FMIN = 2e-4
SENS_FMAX = 1e5
SENS_NFREQ = 100

SENSITIVITY_PARAMETERS = [
    "Negative electrode relative porosity",
    "Positive particle diffusion time scale [s]",
    "Positive electrode electrolyte diffusion time scale [s]",
    "Separator electrolyte diffusion time scale [s]",
    "Positive electrode charge transfer time scale [s]",
    "Series resistance [Ohm]",
    "Positive electrode relative porosity",
    "Cation transference number",
    "Reference electrolyte capacity [A.s]",
    "Positive electrode capacitance [F]",
    "Positive theoretical electrode capacity [As]",
    "Positive electrode relative thickness",
    "Measured cell capacity [A.s]",
]

# Save only PNG figures in a result folder next to this script.
# In normal .py scripts, __file__ exists. In Jupyter / VS Code interactive cells,
# __file__ does not exist, so we fall back to the current working directory.
try:
    SCRIPT_DIR = Path(__file__).resolve().parent
except NameError:
    SCRIPT_DIR = Path.cwd()

SAVE_DIR = SCRIPT_DIR / "result"
SAVE_DIR.mkdir(parents=True, exist_ok=True)

FIG_DPI = 300

# Colormap used for sensitivity curves.
# Good presentation choices: "viridis", "plasma", "cividis", "turbo".
SENSITIVITY_CMAP = "viridis"

# Use a logarithmic color scale because the parameter values are log-spaced.
USE_LOG_COLOR_SCALE = True


# =============================================================================
# HELPER FUNCTIONS
# =============================================================================

def safe_filename(text):
    out = []
    for char in text:
        if char.isalnum():
            out.append(char)
        else:
            out.append("_")
    return "_".join("".join(out).split("_"))


def positive_log_values(base_value, factor, n_values):
    base_value = float(base_value)

    if not np.isfinite(base_value):
        raise ValueError(f"Base value is not finite: {base_value}")

    if base_value <= 0:
        raise ValueError(
            f"Base value must be positive for log-spaced sensitivity: {base_value}"
        )

    return np.logspace(
        np.log10(base_value / factor),
        np.log10(base_value * factor),
        n_values,
    )


def get_parameter_base_value(grouped, parameter_name):
    """
    Return the base value for a grouped sensitivity parameter.

    Special case:
        "Positive theoretical electrode capacity [As]"

    GroupedSPMe does not expose this as an independent parameter. In the
    model equations, the positive theoretical capacity is effectively

        Q_th_p = Measured cell capacity [A.s] /
                 (Maximum positive stoichiometry - Minimum positive stoichiometry)

    Therefore this script treats it as an alias and updates
    "Measured cell capacity [A.s]" during the simulation.
    """
    if parameter_name in grouped:
        return float(grouped[parameter_name])

    if parameter_name == "Positive theoretical electrode capacity [As]":
        y_min = float(grouped["Minimum positive stoichiometry"])
        y_max = float(grouped["Maximum positive stoichiometry"])
        delta_y = y_max - y_min

        if delta_y <= 0:
            raise ValueError(
                "Could not compute Positive theoretical electrode capacity because "
                "Maximum positive stoichiometry <= Minimum positive stoichiometry."
            )

        return float(grouped["Measured cell capacity [A.s]"]) / delta_y

    raise KeyError(parameter_name)


def apply_parameter_value(grouped, parameter_name, value):
    """
    Apply one sensitivity value to a grouped-parameter dictionary.
    """
    grouped = copy.deepcopy(grouped)
    value = float(value)

    if parameter_name in grouped:
        grouped[parameter_name] = value
        return grouped

    if parameter_name == "Positive theoretical electrode capacity [As]":
        y_min = float(grouped["Minimum positive stoichiometry"])
        y_max = float(grouped["Maximum positive stoichiometry"])
        delta_y = y_max - y_min

        grouped["Measured cell capacity [A.s]"] = value * delta_y
        return grouped

    raise KeyError(parameter_name)


def run_one_parameter_sensitivity(grouped_base, parameter_name, frequencies):
    base_value = get_parameter_base_value(grouped_base, parameter_name)
    values = positive_log_values(base_value, FACTOR, N_VALUES)

    Z_all = np.zeros((len(frequencies), len(values)), dtype=complex)

    print("\n" + "=" * 100)
    print(parameter_name)
    print("=" * 100)
    print("Base value:", base_value)
    print("Minimum tested:", values[0])
    print("Maximum tested:", values[-1])

    for j, value in enumerate(values):
        print(f"  {j + 1}/{len(values)}: {parameter_name} = {value:.6e}")

        grouped = apply_parameter_value(grouped_base, parameter_name, value)

        Z = run_grouped_spme_eis(
            grouped,
            frequencies,
            model_options=MODEL_OPTIONS,
            var_pts=VAR_PTS,
        )

        Z_all[:, j] = Z

    return values, Z_all, base_value


def plot_parameter_sensitivity(parameter_name, values, Z_all, base_value, save_dir):
    """
    Presentation-quality Nyquist plot using a colormap instead of a line legend.

    The colorbar shows the varied parameter as:
        parameter value / base value
    """
    fig, ax = plt.subplots(figsize=(8.4, 6.7))

    n_freq = Z_all.shape[0]
    mark_every = max(1, n_freq // 14)

    multipliers = np.asarray(values, dtype=float) / float(base_value)

    cmap = plt.get_cmap(SENSITIVITY_CMAP)

    if USE_LOG_COLOR_SCALE:
        norm = mcolors.LogNorm(
            vmin=float(np.nanmin(multipliers)),
            vmax=float(np.nanmax(multipliers)),
        )
    else:
        norm = mcolors.Normalize(
            vmin=float(np.nanmin(multipliers)),
            vmax=float(np.nanmax(multipliers)),
        )

    for j, value in enumerate(values):
        multiplier = float(multipliers[j])
        is_base = np.isclose(multiplier, 1.0, rtol=1e-4, atol=1e-8)

        ax.plot(
            np.real(Z_all[:, j]),
            -np.imag(Z_all[:, j]),
            "-o",
            color=cmap(norm(multiplier)),
            linewidth=2.8 if is_base else 1.8,
            markersize=4.4 if is_base else 3.0,
            markevery=mark_every,
            alpha=1.0 if is_base else 0.92,
            zorder=3 if is_base else 2,
        )

    # Colorbar replaces the curve legend.
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])

    cbar = fig.colorbar(
        sm,
        ax=ax,
        pad=0.025,
        fraction=0.055,
    )

    cbar.set_label("Parameter / base value", fontsize=12)
    cbar.set_ticks(multipliers)
    cbar.set_ticklabels([f"{m:.2f}×" for m in multipliers])
    cbar.ax.tick_params(labelsize=9)

    ax.text(
        0.02,
        0.98,
        f"Base = {base_value:.3e}",
        transform=ax.transAxes,
        ha="left",
        va="top",
        fontsize=10,
        bbox={
            "boxstyle": "round,pad=0.3",
            "facecolor": "white",
            "edgecolor": "0.7",
            "alpha": 0.85,
        },
    )

    ax.set_xlabel(r"$Z_\mathrm{r}(\omega)$ [$\Omega$]", fontsize=13)
    ax.set_ylabel(r"$-Z_\mathrm{j}(\omega)$ [$\Omega$]", fontsize=13)
    ax.set_title(
        f"GroupedSPMe sensitivity\n{parameter_name}, SOC = {SOC}",
        fontsize=14,
        pad=12,
    )
    ax.grid(True, alpha=0.35)
    ax.tick_params(axis="both", labelsize=11)
    ax.set_aspect("equal", adjustable="box")

    fig.tight_layout()

    file_base = safe_filename(parameter_name)
    png_path = save_dir / f"{file_base}_nyquist_colormap.png"

    fig.savefig(png_path, dpi=FIG_DPI, bbox_inches="tight")

    print("Saved:", png_path)

    # Show plot when running in VS Code.
    plt.show()

    return png_path


def sensitivity_metric(Z_all):
    """
    Simple scalar metric for ranking parameters.

    It compares each curve against the centre/base curve and reports the largest
    relative deviation in |Z| over all frequencies and tested values.
    """
    centre_index = Z_all.shape[1] // 2
    Z_base = Z_all[:, centre_index]

    denom = np.nanmax(np.abs(Z_base))
    if denom == 0 or not np.isfinite(denom):
        return np.nan

    diff = np.abs(Z_all - Z_base[:, None])
    return float(np.nanmax(diff) / denom)


# =============================================================================
# MAIN
# =============================================================================

frequencies = np.logspace(
    np.log10(SENS_FMIN),
    np.log10(SENS_FMAX),
    SENS_NFREQ,
)

params, diagnostics, scalar_grouping_overrides = make_physical_params_from_bpx(
    BPX_FILE,
    soc=SOC,
    contact_resistance_ohm=CONTACT_RESISTANCE_OHM,
)

grouped_base, grouped_metadata = make_grouped_params_from_physical(
    params,
    scalar_grouping_overrides,
    tau_ct_mode=TAU_CT_MODE,
    tau_ct_multiplier=TAU_CT_MULTIPLIER,
    use_total_area_for_grouping=USE_TOTAL_AREA_FOR_GROUPING,
    force_nominal_capacity=FORCE_NOMINAL_CAPACITY_IN_GROUPED_MODEL,
)

summary_rows = []

for parameter_name in SENSITIVITY_PARAMETERS:
    try:
        values, Z_all, base_value = run_one_parameter_sensitivity(
            grouped_base,
            parameter_name,
            frequencies,
        )

        png_path = plot_parameter_sensitivity(
            parameter_name,
            values,
            Z_all,
            base_value,
            SAVE_DIR,
        )

        metric = sensitivity_metric(Z_all)
        status = "completed"

    except Exception as err:
        print("\nSKIPPED:", parameter_name)
        print("Reason:", repr(err))

        base_value = np.nan
        metric = np.nan
        status = f"skipped: {repr(err)}"
        png_path = ""

    summary_rows.append(
        {
            "Parameter": parameter_name,
            "Status": status,
            "Base value": base_value,
            "Sensitivity metric": metric,
            "PNG": str(png_path),
        }
    )

# Summary ranking plot, saved as PNG only and shown.
rank_rows = [
    row for row in summary_rows
    if np.isfinite(row["Sensitivity metric"])
]

rank_rows = sorted(
    rank_rows,
    key=lambda row: row["Sensitivity metric"],
    reverse=True,
)

if len(rank_rows) > 0:
    fig, ax = plt.subplots(figsize=(9.0, max(5.5, 0.42 * len(rank_rows))))

    y_pos = np.arange(len(rank_rows))
    metrics = [row["Sensitivity metric"] for row in rank_rows]
    names = [row["Parameter"] for row in rank_rows]

    ax.barh(y_pos, metrics)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(names, fontsize=9)
    ax.invert_yaxis()
    ax.set_xlabel("Maximum relative change in impedance response", fontsize=12)
    ax.set_title("GroupedSPMe parameter-sensitivity ranking", fontsize=14, pad=12)
    ax.grid(True, axis="x", alpha=0.35)
    fig.tight_layout()

    ranking_png = SAVE_DIR / "sensitivity_ranking.png"
    fig.savefig(ranking_png, dpi=FIG_DPI, bbox_inches="tight")

    print("Saved:", ranking_png)

    plt.show()

print("\nDONE")
print("Output folder:", SAVE_DIR)
print("Only PNG figures were saved. No PDF, CSV, or MAT files were written.")
```
